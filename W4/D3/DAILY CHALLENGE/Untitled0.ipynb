```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.ndimage import rotate

# Initialize a list to store rotated images
rotated_images = []

# Set up the figure for displaying images
plt.figure(figsize=(20, 20))

# Loop through the first 5 images in the 'flowers/flowers' directory
image_dir = 'flowers/flowers'
for i, file in enumerate(os.listdir(image_dir)[:5]):  # Use [:5] to limit to the first 5 images
    # Get the image path
    image_path = os.path.join(image_dir, file)

    # Read the image
    img = mpimg.imread(image_path)

    # Rotate the image by 30 degrees
    rotated_img = rotate(img, 30, reshape=False)

    # Store the rotated image
    rotated_images.append(rotated_img)

    # Display the rotated image
    ax = plt.subplot(3, 5, i + 1)  # Create a subplot
    ax.title.set_text(file)        # Set the title as the file name
    plt.imshow(rotated_img)        # Show the rotated image
    plt.axis('off')                # Hide axes for a cleaner look

# Show the plot
plt.tight_layout()
plt.show()
```
