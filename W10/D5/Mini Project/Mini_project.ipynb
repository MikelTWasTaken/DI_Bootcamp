Data Pre-processing and cleaning
- Remove null rows/ columns: 
    - COL - Job Title Additional Info, Minimum Pay, Maximum Pay, Pay Rate... 
    - ROW - Number of ApplicantsNumber of Applicants, Company Name, Company Industry, Company Size  
- impute numerical values where possible
- Create new columns for columns containing lots of info (job location, Job Skills (one hot encoding))
- Save to new csv

```python
df['Job Skills'] = df['Job Skills'].fillna('')

# Create the skills list (split the 'Job Skills' column by commas)
skills_list = df['Job Skills'].str.split(', ')

# Iterate over each skill in the unique list of skills
for skill in set([skill for sublist in skills_list for skill in sublist]):
    # Create a new column with 1 if the skill is in the 'Job Skills', else 0
    df[skill] = df['Job Skills'].apply(lambda x: 1 if skill in x else 0)

# Print the final DataFrame with individual skill columns
print(df)
```

```python
# drop the 'job skill' column
df = df.drop(columns=['Job Skills'])

```

```python
#save to a new csv
df.to_csv('cleaned job data.csv', index=False)
```
